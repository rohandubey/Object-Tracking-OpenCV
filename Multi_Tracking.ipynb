```python
import cv2
import numpy as np
from random import randint

tracker_types = ['BOOSTING','MIL','KCF','TLD','MEDIANFLOW','GOTURN','MOSSE','CSRT']

def tracker_name(tracker_type):
    if tracker_type ==tracker_types[0]:
        tracker = cv2.TrackerBoosting_create()
    elif tracker_type ==tracker_types[1]:
        tracker = cv2.TrackerMIL_create()
    elif tracker_type ==tracker_types[2]:
        tracker = cv2.TrackerKCF_create()
    elif tracker_type ==tracker_types[3]:
        tracker = cv2.TrackerTLD_create()
    elif tracker_type ==tracker_types[4]:
        tracker = cv2.TrackerMedianFlow_create()
    elif tracker_type ==tracker_types[5]:
        tracker = cv2.TrackerGOTURN_create()
    elif tracker_type ==tracker_types[6]:
        tracker = cv2.TrackerMOSSE_create()
    elif tracker_type ==tracker_types[7]:
        tracker = cv2.TrackerCSRT_create()
    else:
        tracker =None
        print('No tracker Found')
        print('Chose from these Tracker')
        i=0
        for tr in tracker_types:
            i+=1
            print(i,' ',tr)
    return tracker
if __name__ =='__main__':
    print('default tracking algorithm Moose \nAvailable algorithms are:\n')
    i=0
    for tr in tracker_types:
        i+=1
        print(i,' ',tr)
    tracker_type = 'MOOSE'
    cap = cv2.VideoCapture('Vehicles.mp4')
    ret, frame = cap.read()
    if  not ret:
        print('Cannot read the Video')
    rects = []
    color = []
    while True:
        rects_box = cv2.selectROI('MultiTracker',frame)
        rects.append(rects_box)
        color.append((randint(64,255),randint(64,255),randint(64,255)) )
                      
        print('press q to stop selecting boxes and Multi Tracking')
        print('press any key to select another box')
        
        if cv2.waitKey(0) & 0xFF ==113: #113 = q
            break
    print(f'selected boxes {rects}')
    multitracker = cv2.MultiTracker_create()
    
    for rect_box in rects:
        multitracker.add(tracker_name(tracker_type),frame,rect_box)
    
    while cap.isOpened():
        ret,frame =cap.read()
        if not ret:
            break
        ret , boxes = multitracker.update(frame)
        for (i,newbox) in enumerate(boxes):
            pts1 = (int(newbox[0]),int(newbox[1]))
            pts2 = (int(newbox[0])+int(newbox[2]),int(newbox[1])+int(newbox[3]))
            cv2.rectangle(frame,pts1,pts2,color[i],2,1)
        cv2.imshow('MUltiTracker',frame)
        if cv2.waitKey(300) & 0xFF == 27:
            break
cap.release()
cv2.destroyAllWindows()
```
